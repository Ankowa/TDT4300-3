```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cdist

MINPOINTS = 2
EPS = 3

points = pd.read_csv("dbscan_data.csv")
coords = ["x", "y"]
points.rename(columns={"id": "point"}, inplace=True)
points.head()
```

```python
def get_distance_matrix(data: pd.DataFrame) -> pd.DataFrame:
    df_array = data[coords].to_numpy()
    dist_mat = cdist(df_array, df_array)
    return pd.DataFrame(dist_mat, columns=data.point, index=data.point)
```

```python
dist = get_distance_matrix(points)
```

```python
n = dist.apply(lambda x: x <= EPS, axis=1).sum(axis=1)
```

```python
n[n < MINPOINTS].index.tolist()
```

```python
core = n[n >= MINPOINTS].index.tolist()
core
```

```python
corepoints = dist.loc[core, core]
```

```python
mask = corepoints.apply(lambda x: x <= EPS, axis=1)
```

```python
cn = mask.apply(lambda x: mask.columns[x].tolist(), axis=1)
cn
```

```python
clusters = list()
for row in cn.items():
    cores = list(set([row[0]] + row[1]))
    if len(clusters) == 0:
        clusters.append(cores)
    else:
        isnew = True
        for idx, cluster in enumerate(clusters):
            if len(set(cluster).intersection(set(cores))) != 0:
                cluster = list(set(cluster).union(set(cores)))
                clusters[idx] = cluster
                isnew = False
                break
        if isnew:
            clusters.append(cores)
```
